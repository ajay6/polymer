```python
import numpy as np
no_of_ch = 278
natoms=24000
nbonds=5282
system_size = 20.0
i=0
x0=0
y0=0
z0=0
no_of_bt = 20
chain1 = np.zeros(shape=(no_of_bt,3))
while(i<no_of_bt):
    chain1[i,0]=x0
    chain1[i,1]=y0
    chain1[i,2]=z0
    x0=x0+0.85
    i+=1
#print('chain 1',chain1)
chains = np.zeros(shape=(no_of_ch,no_of_bt,3))
#print('chains',chains)
ind1 = 0

while(ind1<no_of_bt):
    #print('ind1',ind1)
    ind2 = 0
    while(ind2<3):
        #print('ind2',ind2)
        #print('chain1 elem',chain1[ind1,ind2])
        chains[0,ind1,ind2]=chain1[ind1,ind2]
        ind2+=1
    ind1+=1
ch=1
#print('chains',chains)
ran = []
while(ch<no_of_ch):
    ran=(np.random.rand(3)*20)
    ind1=0
    while ind1<no_of_bt:
        ind2=0
        while ind2<3:
            chains[ch,ind1,ind2] = chain1[ind1,ind2]+ran[ind2]
            ind2+=1
        ind1+=1
    ch=ch+1
#print(chains)
with open('ajay.txt', "w") as f:
    f.write('#Random atoms - Generated for LAMMPS SCRIPT FILE\n\n')
    f.write('{} atoms\n'.format(natoms)) 
    f.write('{} bonds\n'.format(nbonds)) 
    f.write('\n')
    f.write('{} atom types\n'.format(2))
    f.write('\n')
    f.write('{} bond types\n'.format(1))
    f.write('\n')
    f.write('{} {} xlo xhi\n'.format(0.0, system_size))  
    f.write('{} {} ylo yhi\n'.format(0.0, system_size)) 
    f.write('{} {} zlo zhi\n'.format(0.0, system_size)) 
    f.write('\n')

    f.write('Atoms\n\n')
    
    SN=1
    i=1
    while i<no_of_ch+1:
        j=1
        while j<no_of_bt+1:
            f.write('{} {} {} {} {} {}\n'.format(SN,0,i,chains[i-1,j-1,0],chains[i-1,j-1,1],chains[i-1,j-1,2]))
            j=j+1
            SN=SN+1
        i=i+1


    it=5561
    while it<24001:
        ran = np.random.rand(3)*no_of_bt

        f.write('{} {} {} {} {} {}\n'.format(SN, 0, 0,ran[0],ran[1],ran[2]))
        it=it+1
        SN+=1
        
    f.write('\nBonds\n\n')
    SN=1
    i=1
    Bond_no=1
    while i<no_of_ch+1:
        j=1
        while j<no_of_bt:
            f.write('{} {} {} {}\n'.format(SN,i,Bond_no,Bond_no+1))
            j+=1
            SN=SN+1
            Bond_no+=1
        i+=1
        Bond_no+=1
```
